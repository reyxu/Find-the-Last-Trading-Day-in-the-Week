```python
from datetime import datetime,timedelta,date,time
import pandas as pd

# You will need the following package to do it 
!pip3 install pandas_market_calendars
import pandas_market_calendars as mcal
```

## Step 1: set date range 

```python
start = '2001-01-01'
end = '2020-06-21'

dates = pd.DataFrame({'date':pd.date_range(start, end,freq="D")})
dates = dates.set_index('date')
dates
```

## Step 2: find trading days within that selected date range

```python
nyse = mcal.get_calendar('NYSE')
trading_days = nyse.schedule(start_date = start, end_date = end) 
trading_days = trading_days.reset_index()
trading_days = trading_days.drop(columns = ['market_open', 'market_close'])
trading_days['trading days'] = pd.to_datetime(trading_days['index'], format = '%Y-%m-%d')
trading_days = trading_days.set_index('index')
trading_days
```

## Step 3: concat two data frames by index and drop index

```python
result = pd.concat([dates, trading_days], axis=1)
result.reset_index(drop=True, inplace =True)
result
```

## Step 4: use resample to find the last trading day of every week

```python
last_trading_day = result.resample('W', on='trading days').last()
last_trading_day.reset_index(drop = True, inplace = True)
last_trading_day = last_trading_day.rename(columns={"trading days": "Last Trading Day of the Week"})
last_trading_day['Year'] = last_trading_day['Last Trading Day of the Week'].dt.year
last_trading_day['Week Number'] = last_trading_day['Last Trading Day of the Week'].dt.week
last_trading_day['Day'] = last_trading_day['Last Trading Day of the Week'].dt.day_name()
last_trading_day.head(50)
```
